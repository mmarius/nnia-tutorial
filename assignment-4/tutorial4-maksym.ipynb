#### 1. Sigmoid Function

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def dsigmoid(x):
    return (sigmoid(x) * (1.0 - sigmoid(x)))

xs = np.linspace(-5.0, 5.0, 100)

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.axhline(0, ls='--', c='black', lw=.5)
ax1.axvline(0, ls='--', c='black', lw=.5)
ax2.axhline(0, ls='--', c='black', lw=.5)
ax2.axvline(0, ls='--', c='black', lw=.5)
ax1.plot(xs, sigmoid(xs), c='red')
ax1.plot(xs, sigmoid(-xs), c='purple', ls='--')

ax2.plot(xs, dsigmoid(xs), c='blue')
ax1.set_title('Sigmoid')
ax2.set_title('Derivative');
```

#### 5. Multiple Linear Regression

```python
from sklearn.datasets import make_regression
```

```python
# Generate data
X = np.linspace(0, 10, 30)
y = np.sin(X).ravel()

# Add noise to targets
y = np.asarray([y[i] + 0.2 * np.random.normal(0.0, 1.0) for i in range(len(y))])
```

```python
print(X.shape)
print(y.shape)
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
ax.set_title('Dummy Data');
```

```python
xs = np.linspace(0, 10, 100)
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y, label='Training data')
for i in [1, 3, 9, 13]:
    # Let's fit a polynomial
    z = np.polyfit(X, y, i)
    p = np.poly1d(z)
    # Compute MSE on training data
    print('d = {:d} MSE = {:.3f}'.format(i, np.mean((y - p(X))**2)))
    
    # Plot the fitted polynomial
    ax.plot(xs, p(xs), label='Degree {:d} polynomial'.format(i))
    
ax.legend(bbox_to_anchor=(1.5,1.01));
```
